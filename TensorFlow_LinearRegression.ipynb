```python
import tensorflow as tf
import numpy as np
```

```python
#initialize the value
a = tf.Variable(tf.constant(5.0))
b = tf.Variable(tf.constant(5.0))
x_data = tf.placeholder(tf.float32)
y_val = 75
sess = tf.Session()
lr = 0.05
```

```python
#construct the hypothesis
mul = tf.add(tf.multiply(a,x_data),b)

#loss
loss = tf.square(tf.subtract(mul , y_val))
```

```python
init = tf.global_variables_initializer()
sess.run(init)

#optimizer
opt = tf.train.GradientDescentOptimizer(lr)
train_step = opt.minimize(loss)
```

```python
for i in range(0,25):
    print('{}'.format(sess.run(a)))
    print('{}'.format(sess.run(b)))
    sess.run(train_step , feed_dict={x_data:2.5})
    print('updated a {}'.format(sess.run(a)))
    print('updated b {}'.format(sess.run(b)))
    print('ouput value {}'.format(sess.run(mul , feed_dict={x_data:2.5})))
```
